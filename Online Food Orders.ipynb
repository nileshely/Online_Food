```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/online-food-dataset/onlinefoods.csv')
```

```python
# Age distribution
plt.figure(figsize=(14, 10))
plt.subplot(3, 2, 1)
sns.histplot(data=data, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 2)
sns.countplot(data=data, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
# Marital status distribution
plt.figure(figsize=(12, 6))
plt.subplot(3, 2, 3)
sns.countplot(data=data, x='Marital Status')
plt.title('Marital Status Distribution')
plt.show()
```

```python
# Occupation distribution
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 4)
sns.countplot(data=data, y='Occupation', order=data['Occupation'].value_counts().index)
plt.title('Occupation Distribution')
plt.show()
```

```python
# Monthly income distribution
plt.figure(figsize=(20,8))
plt.subplot(3, 2, 5)
sns.histplot(data=data, x='Monthly Income', bins=20, kde=True)
plt.title('Monthly Income Distribution')
plt.show()
```

```python
# Educational qualifications distribution
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 6)
sns.countplot(data=data, y='Educational Qualifications', order=data['Educational Qualifications'].value_counts().index)
plt.title('Educational Qualifications Distribution')
plt.show()
```

```python
# Current status of orders
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Output')
plt.title('Order Status')
plt.show()
```

```python
# Feedback distribution
plt.subplot(1, 2, 2)
sns.countplot(data=data, x='Feedback')
plt.title('Feedback Distribution')
plt.show()
```
